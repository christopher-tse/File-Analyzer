### Step 2a

I will need to represent the genre of the video games and the North America sales for these video games as information in order to create the visualization that I'm looking for. It is important to represent these pieces of information because for a given North America sales I will need to know the corresponding genre. A compound for each genre's sale and an arbitrary number of genre sales will be needed. I will also use Enumeration to show the different kinds of genre.

````python
from cs103 import *
from typing import NamedTuple
from typing import List
from enum import Enum

Genre = Enum('Genre', ['Sports', 'Platform', 'Racing', 'Role_Playing', 'Puzzle', 'Misc', 'Shooter', 'Simulation',
                       'Action', 'Fighting', 'Adventure', 'Strategy'])
# interp. a video game can have the genre Sports, Platform, Racing, Role-Playing, Puzzle, Misc, Shooter,
# Simulation, Action, Fighting, Adventure, or Strategy
# Examples are redundant for Enumerations

# template based on One Of (12 cases) and Atomic-Distinct (12x)
def fn_for_genre(g: Genre) -> ...:
    if g == Genre.Sports:
        return ...
    elif g == Genre.Platform:
        return ...
    elif g == Genre.Racing:
        return ...
    elif g == Genre.Role_Playing:
        return ...
    elif g == Genre.Puzzle:
        return ...
    elif g == Genre.Misc:
        return ...
    elif g == Genre.Shooter:
        return ...
    elif g == Genre.Simulation:
        return ...
    elif g == Genre.Action:
        return ...
    elif g == Genre.Fighting:
        return ...
    elif g == Genre.Adventure:
        return ...
    elif g == Genre.Strategy:
        return ...
    
GenreSale = NamedTuple('GenreSale', [('genre', Genre),
                                      ('na_sale', float)]) # in range [0.0, ...)
# interp. a video game has a genre and an associated North American sale ('na_sale') in millions of dollars

GS1 = GenreSale(Genre.Sports, 41.9)
GS2 = GenreSale(Genre.Platform, 30.4)
GS3 = GenreSale(Genre.Racing, 26.5)
GS4 = GenreSale(Genre.Role_Playing, 20.4)
GS5 = GenreSale(Genre.Puzzle, 19.02)
GS6 = GenreSale(Genre.Misc, 17)
GS7 = GenreSale(Genre.Shooter, 13.89)
GS8 = GenreSale(Genre.Simulation, 12.7)
GS9 = GenreSale(Genre.Action, 8.5)
GS10 = GenreSale(Genre.Fighting, 4.6)
GS11 = GenreSale(Genre.Adventure, 2.1)
GS12 = GenreSale(Genre.Strategy, 0.0)

# template based on Compound (2 fields) and Reference Rule
def fn_for_genre_sale(gs: GenreSale) -> ...:
    return ...(fn_for_genre(gs.genre),
               gs.na_sale)

# List[GenreSale]
# interp. a list of genre sales

LGS0 = []
LGS1 = [GS1]
LGS2 = [GS1, GS2, GS3, GS4, GS5]
LGS3 = [GS7, GS1, GS9, GS4, GS2, GS11]

# template based on Arbitrary-Sized and Reference Rule
def fn_for_logs(logs: List[GenreSale]) -> ...:
    # description for the acc
    acc = ... # type: ...
    for gs in logs:
        acc = ...(fn_for_genre_sale(gs), acc)
    return ...(acc)
````

````python
# Analysis Program Template
from cs103 import *
from typing import NamedTuple, List
import csv
import matplotlib.pyplot as pyplot

def main(filename: str) -> None:
    """
    Reads the file from given filename and displays a bar chart comparing the total North America sales of 
    different video game genres, then returns None
    """
    # template based on function composition
    return bar_chart_sales_vs_genre(read(filename))

def parse_genre(g: str) -> Genre:
    """
    takes a given string (which must be one of the following - Sports, Platform, Racing, Role-Playing, Puzzle,
    Misc, Shooter, Simulation, Action, Fighting, Adventure, or Strategy) and returns the corresponding Genre
    """
    #return Genre.Sports #body of stub
    #template from Atomic Non-Distinct
    
    if g == 'Sports':
        return Genre.Sports
    elif g == 'Platform':
        return Genre.Platform
    elif g == 'Racing':
        return Genre.Racing
    elif g == 'Role-Playing':
        return Genre.Role_Playing
    elif g == 'Puzzle':
        return Genre.Puzzle
    elif g == 'Misc':
        return Genre.Misc
    elif g == 'Shooter':
        return Genre.Shooter
    elif g == 'Simulation':
        return Genre.Simulation
    elif g == 'Action':
        return Genre.Action
    elif g == 'Fighting':
        return Genre.Fighting
    elif g == 'Adventure':
        return Genre.Adventure
    elif g == 'Strategy':
        return Genre.Strategy

def read(filename: str) -> List[GenreSale]:
    """    
    reads information from the specified file and returns a list of genre sales
    """
    # logs contains the result so far
    logs = [] # type: List[GenreSale]

    with open(filename, encoding = "ISO-8859-1") as csvfile:
        
        reader = csv.reader(csvfile)
        next(reader) # skip header line

        for row in reader:
            genre = parse_genre(row[4])
            sales = parse_float(row[6])
            
            gs = GenreSale(genre, sales)
            logs.append(gs)
    
    return logs

def bar_chart_sales_vs_genre(logs: List[GenreSale]) -> None:
    """
    displays a bar chart comparing the total North America sales of different video game genres,
    returns None
    """
    #return None #body of stub
    
    #sports = get_sports_sales(logs)
    #platform = get_platform_sales(logs)
    #racing = get_racing_sales(logs)
    #role_play = get_role_sales(logs)
    #puzzle = get_puzzle_sales(logs)
    #misc = get_misc_sales(logs)
    #shooter = get_shooter_sales(logs)
    #simulation = get_sim_sales(logs)
    #action = get_action_sales(logs)
    #fighting = get_fight_sales(logs)
    #adventure = get_adven_sales(logs)
    #strategy = get_strat_sales(logs)
    
    sports = get_sales(logs, Genre.Sports)
    platform = get_sales(logs, Genre.Platform)
    racing = get_sales(logs, Genre.Racing)
    role_play = get_sales(logs, Genre.Role_Playing)
    puzzle = get_sales(logs, Genre.Puzzle)
    misc = get_sales(logs, Genre.Misc)
    shooter = get_sales(logs, Genre.Shooter)
    simulation = get_sales(logs, Genre.Simulation)
    action = get_sales(logs, Genre.Action)
    fighting = get_sales(logs, Genre.Fighting)
    adventure = get_sales(logs, Genre.Adventure)
    strategy = get_sales(logs, Genre.Strategy)
    
    genre_sales = [sports, platform, racing, role_play, puzzle, misc, shooter, simulation, action,
                   fighting, adventure, strategy]
    
    bar_width = 40
    
    left_side_bar_spacing = [5, 55, 105, 155, 205, 255, 305, 355, 405, 455, 505, 555]
    
    opacity = 0.9
    
    bars = pyplot.bar(left_side_bar_spacing,
                      genre_sales,
                      bar_width,
                      alpha = opacity,
                      align = 'edge',
                      color = ['b', 'g', 'r', 'c', 'm', 'y', 'k', (0.76, 0.83, 0.1), (0.123, 0.5, 0.3), 
                               (0.8, 0.75, 0.22), (0.77, 0.96, 0.3), (0.55, 0.41, 1.0)])
    
    pyplot.xlabel('Video Game Genre')
    pyplot.ylabel('Total North America Sales (in millions of dollars)')
    pyplot.title('Total North America Sales of Video Game Genres')
    
    x_cord_labels = [25, 75, 125, 175, 225, 275, 325, 375, 425, 475, 525, 575]
    
    tick_labels = ['Sports', 'Platform', 'Racing', 'Role-Playing', 'Puzzle', 'Misc', 'Shooter', 'Simulation',
                   'Action', 'Fighting', 'Adventure', 'Strategy']
    
    pyplot.xticks(x_cord_labels, tick_labels, rotation='vertical')
    
    pyplot.show()
    
    return None

def get_sales(logs: List[GenreSale], gr: Genre) -> float:
    """
    returns the total sales for specified genres
    """
    #return 0.0 #body of stub
    #template from List[GenreSale]
    
    # total stores the total north america sales for the genre seen so far
    total = 0.0 # type: float
    for g in logs:
        if g.genre == gr:
            total = total + g.na_sale
    return total
    
    
# Begin testing
start_testing()

# Examples and tests for main
expect(main("global_video_game_sales_in_millions.csv"), None)
expect(main("global_video_game_sales_in_millions_test1.csv"), None)
expect(main("global_video_game_sales_in_millions_test2.csv"), None)


# Examples and tests for parse_genre
expect(parse_genre('Sports'), Genre.Sports)
expect(parse_genre('Platform'), Genre.Platform)
expect(parse_genre('Racing'), Genre.Racing)
expect(parse_genre('Role-Playing'), Genre.Role_Playing)
expect(parse_genre('Puzzle'), Genre.Puzzle)
expect(parse_genre('Misc'), Genre.Misc)
expect(parse_genre('Shooter'), Genre.Shooter)
expect(parse_genre('Simulation'), Genre.Simulation)
expect(parse_genre('Action'), Genre.Action)
expect(parse_genre('Fighting'), Genre.Fighting)
expect(parse_genre('Adventure'), Genre.Adventure)
expect(parse_genre('Strategy'), Genre.Strategy)


# Examples and tests for read
expect(read("global_video_game_sales_in_millions_test1.csv"), [GenreSale(Genre.Sports, 41.49),
                                                               GenreSale(Genre.Platform, 29.08),
                                                               GenreSale(Genre.Racing, 15.85),
                                                               GenreSale(Genre.Sports, 15.75)])

expect(read("global_video_game_sales_in_millions_test2.csv"), [GenreSale(Genre.Action, 1.48),
                                                               GenreSale(Genre.Misc, 1.67)])


# Examples and tests for bar_chart_sales_vs_genre
expect(bar_chart_sales_vs_genre(LGS1), None)
expect(bar_chart_sales_vs_genre(LGS3), None)


# Examples and tests for get_sports_sales
expect(get_sales([GenreSale(Genre.Sports, 123), GenreSale(Genre.Shooter, 12), 
                        GenreSale(Genre.Action, 16)], Genre.Sports), 123)
expect(get_sales([GenreSale(Genre.Sports, 123), GenreSale(Genre.Misc, 12), 
                        GenreSale(Genre.Sports, 16.7)], Genre.Sports), 139.7)
expect(get_sales([GenreSale(Genre.Action, 123), GenreSale(Genre.Misc, 12), 
                        GenreSale(Genre.Role_Playing, 16.7)], Genre.Sports), 0.0)


# Examples and tests for get_platform_sales
expect(get_sales([GenreSale(Genre.Sports, 123), GenreSale(Genre.Misc, 12), 
                        GenreSale(Genre.Sports, 16.7)], Genre.Platform), 0.0)
expect(get_sales([GenreSale(Genre.Platform, 123), GenreSale(Genre.Misc, 12), 
                        GenreSale(Genre.Sports, 16.7)], Genre.Platform), 123)
expect(get_sales([GenreSale(Genre.Sports, 123), GenreSale(Genre.Platform, 12), 
                        GenreSale(Genre.Platform, 16.7)], Genre.Platform), 28.7)


# Examples and tests for get_racing_sales
expect(get_sales([GenreSale(Genre.Sports, 123), GenreSale(Genre.Misc, 12), 
                        GenreSale(Genre.Sports, 16.7)], Genre.Racing), 0.0)
expect(get_sales([GenreSale(Genre.Racing, 2.5), GenreSale(Genre.Misc, 12), 
                        GenreSale(Genre.Sports, 16.7)], Genre.Racing), 2.5)
expect(get_sales([GenreSale(Genre.Racing, 123), GenreSale(Genre.Misc, 12), 
                        GenreSale(Genre.Racing, 16.7)], Genre.Racing), 139.7)


# Examples and tests for get_role_sales
expect(get_sales([GenreSale(Genre.Sports, 123), GenreSale(Genre.Misc, 12), 
                        GenreSale(Genre.Action, 16.7)], Genre.Role_Playing), 0.0)
expect(get_sales([GenreSale(Genre.Role_Playing, 12.4), GenreSale(Genre.Misc, 12), 
                        GenreSale(Genre.Action, 16.7)], Genre.Role_Playing), 12.4)
expect(get_sales([GenreSale(Genre.Role_Playing, 12.4), GenreSale(Genre.Role_Playing, 12), 
                        GenreSale(Genre.Action, 16.7)], Genre.Role_Playing), 24.4)


# Examples and tests for get_puzzle_sales
expect(get_sales([GenreSale(Genre.Role_Playing, 12.4), GenreSale(Genre.Misc, 12), 
                        GenreSale(Genre.Action, 16.7)], Genre.Puzzle), 0.0)
expect(get_sales([GenreSale(Genre.Role_Playing, 12.4), GenreSale(Genre.Puzzle, 34), 
                        GenreSale(Genre.Action, 16.7)], Genre.Puzzle), 34)
expect(get_sales([GenreSale(Genre.Role_Playing, 12.4), GenreSale(Genre.Puzzle, 34), 
                        GenreSale(Genre.Puzzle, 16.7)], Genre.Puzzle), 50.7)


# Examples and tests for get_misc_sales
expect(get_sales([GenreSale(Genre.Role_Playing, 12.4), GenreSale(Genre.Puzzle, 34), 
                        GenreSale(Genre.Action, 16.7)], Genre.Misc), 0.0)
expect(get_sales([GenreSale(Genre.Role_Playing, 12.4), GenreSale(Genre.Misc, 13), 
                        GenreSale(Genre.Action, 16.7)], Genre.Misc), 13)
expect(get_sales([GenreSale(Genre.Misc, 12.4), GenreSale(Genre.Puzzle, 34), 
                        GenreSale(Genre.Misc, 16.7)], Genre.Misc), 29.1)


# Examples and tests for get_shooter_sales
expect(get_sales([GenreSale(Genre.Shooter, 17.2), GenreSale(Genre.Adventure, 13.1), 
                        GenreSale(Genre.Role_Playing, 16)], Genre.Shooter), 17.2)
expect(get_sales([GenreSale(Genre.Shooter, 11.2), GenreSale(Genre.Shooter, 19.9), 
                        GenreSale(Genre.Role_Playing, 16)], Genre.Shooter), 31.1)
expect(get_sales([GenreSale(Genre.Role_Playing, 11.2), GenreSale(Genre.Action, 19.9), 
                        GenreSale(Genre.Role_Playing, 16)], Genre.Shooter), 0.0)


# Examples and tests for get_sim_sales
expect(get_sales([GenreSale(Genre.Role_Playing, 12.4), GenreSale(Genre.Puzzle, 34), 
                        GenreSale(Genre.Action, 16.7)], Genre.Simulation), 0.0)
expect(get_sales([GenreSale(Genre.Simulation, 40.4), GenreSale(Genre.Puzzle, 34), 
                        GenreSale(Genre.Action, 16.7)], Genre.Simulation), 40.4)
expect(get_sales([GenreSale(Genre.Simulation, 40.4), GenreSale(Genre.Puzzle, 34), 
                        GenreSale(Genre.Simulation, 16.7)], Genre.Simulation), 57.1)


# Examples and tests for get_action_sales
expect(get_sales([GenreSale(Genre.Action, 30.6), GenreSale(Genre.Adventure, 11.1), 
                        GenreSale(Genre.Role_Playing, 86)], Genre.Action), 30.6)
expect(get_sales([GenreSale(Genre.Shooter, 17.2), GenreSale(Genre.Action, 100.1), 
                        GenreSale(Genre.Action, 160.1)], Genre.Action), 260.2)
expect(get_sales([GenreSale(Genre.Shooter, 17.2), GenreSale(Genre.Adventure, 100.1), 
                        GenreSale(Genre.Misc, 160.1)], Genre.Action), 0.0)


# Examples and tests for get_fight_sales
expect(get_sales([GenreSale(Genre.Simulation, 40.4), GenreSale(Genre.Puzzle, 34), 
                        GenreSale(Genre.Action, 16.7)], Genre.Fighting), 0.0)
expect(get_sales([GenreSale(Genre.Fighting, 41.4), GenreSale(Genre.Puzzle, 34), 
                        GenreSale(Genre.Action, 16.7)], Genre.Fighting), 41.4)
expect(get_sales([GenreSale(Genre.Fighting, 41.4), GenreSale(Genre.Puzzle, 34), 
                        GenreSale(Genre.Fighting, 1.7)], Genre.Fighting), 43.1)


# Examples and tests for get_adven_sales
expect(get_sales([GenreSale(Genre.Fighting, 41.4), GenreSale(Genre.Puzzle, 34), 
                        GenreSale(Genre.Action, 16.7)], Genre.Adventure), 0.0)
expect(get_sales([GenreSale(Genre.Fighting, 41.4), GenreSale(Genre.Puzzle, 34), 
                        GenreSale(Genre.Adventure, 16.7)], Genre.Adventure), 16.7)
expect(get_sales([GenreSale(Genre.Fighting, 41.4), GenreSale(Genre.Adventure, 34), 
                        GenreSale(Genre.Adventure, 16.7)], Genre.Adventure), 50.7)


# Examples and tests for get_strat_sales
expect(get_sales([GenreSale(Genre.Fighting, 41.4), GenreSale(Genre.Adventure, 34), 
                        GenreSale(Genre.Adventure, 16.7)], Genre.Strategy), 0.0)
expect(get_sales([GenreSale(Genre.Fighting, 41.4), GenreSale(Genre.Adventure, 34), 
                        GenreSale(Genre.Strategy, 7.7)], Genre.Strategy), 7.7)
expect(get_sales([GenreSale(Genre.Strategy, 41.4), GenreSale(Genre.Adventure, 34), 
                        GenreSale(Genre.Strategy, 7.7)], Genre.Strategy), 49.1)

# show testing summary
summary()
````
